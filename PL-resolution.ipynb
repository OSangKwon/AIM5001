```python
def isEmpty(s):
    if len(s) == 0:
        return True
    return False
def negation(c):
    if c[0] == '-':
        return c[1:]
    else:
        return '-' + c
def deep_copy(list1):
  list2 = list()
  for i in list1:
    list2.append(i)
  return list2
```

```python
def PLResolve(ci,cj):
    copy_ci = deep_copy(ci)
    copy_cj = deep_copy(cj)
    if len(copy_ci) == 1 and len(copy_cj) == 1:
        if copy_ci[0] == negation(copy_cj[0]):
            return 'end'
    for i in copy_ci:
        if negation(i) in copy_cj:
            copy_ci.remove(i)
            copy_cj.remove(negation(i))
            return set(copy_ci+copy_cj)
    return 'pass'
```

```python
def PLResolution(KB,alpha):
    #KB : the knowledge base
    clauses = deep_copy(KB)
    clauses.append(alpha)
    while True:
        new_clauses = list()
        n_clauses = len(clauses)
        for i in range(0,n_clauses-1):
            for j in range(i+1,n_clauses):
                resolvents = PLResolve(clauses[i],clauses[j])
                if resolvents == 'end':
                    return True
                elif resolvents == 'pass':
                    continue
                else:
                    if (resolvents in new_clauses) or (resolvents in clauses):
                        continue
                    else:
                        new_clauses.append(resolvents)
        if isEmpty(new_clauses):
            return False
        clauses += new_clauses
```

```python
KB3 = [ {'P12'}]
KB2 = [ {'P12', 'B11'}, {'-B11'} ]
KB1 = [ {'-P21', 'B11'}, {'-B11', 'P12', 'P21'}, {'-P12', 'B11'}, {'-B11'} ]
KB4 = [ {'-P'}, {'R'}, {'P', 'T'} ]
KB5 = [ {'-q', 'p'}, {'-p', 'r'}, {'q', 's'}, {'-s'} ]
alpha1 = {'-P12'}
alpha2 = {'P12'}
alpha3 = {'P12', 'B11'}
alpha4 = {'-T'}
_alpha4 = {'T'}
alpha5 = {'-r'}
_alpha5 = {'r'}
```

```python
print(PLResolution(KB3, alpha1)) # True
print(PLResolution(KB2, alpha1)) # True
print(PLResolution(KB2, alpha2)) # False
print(PLResolution(KB1, alpha2)) # True
print(PLResolution(KB2, alpha3)) # False
print(PLResolution(KB4, alpha4)) # True
print(PLResolution(KB4, _alpha4)) # False
print(PLResolution(KB5, alpha5)) # True
print(PLResolution(KB5, _alpha5)) # False
```
